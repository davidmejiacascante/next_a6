```python
import pandas as pd
import numpy as np
import json
from glob import glob
```

```python
# adjust the pattern as needed, e.g., "data/*.csv"
files = glob("data/*.csv")

df = pd.concat((pd.read_csv(f) for f in files), ignore_index=True)
```

```python
df['Timestamp [UTC]'] = pd.to_datetime(df['Timestamp [UTC]'], errors='coerce')
df['Event'] = df['Event'].fillna('NA').astype('string')
df['Server'] = df['Server'].fillna('NA').astype('string')
df['Application'] = df['Application'].fillna('NA').astype('string')
df['Properties'] = df['Properties'].apply(json.dumps)
```

```python
df.info()
```

```python
df.to_parquet('data/master.parquet')
```
